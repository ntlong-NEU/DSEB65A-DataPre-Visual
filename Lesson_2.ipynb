```python
url1 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv'
url2 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv'
url3 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv'

pop = pd.read_csv(url1)
areas = pd.read_csv(url2)
abbrevs = pd.read_csv(url3)
```

```python
pop.head()
```

```python
areas.head()
```

```python
abbrevs.head()
```

```python
data1 = pd.merge(areas,abbrevs,how = 'outer')
data1.head()
```

```python
data2 = pd.merge(data1,pop, how = 'outer', left_on = 'abbreviation', right_on = 'state/region')
data2.head()
```

```python
data2.drop(columns = ['abbreviation','state/region'],inplace = True)
data2.head()
```

```python
data2.info()
```

```python
data2.isnull().sum()
```

```python
data2.dropna(inplace = True)
data2.isnull().sum()
```

```python
data2.head()
```

```python
data2['ages'].value_counts()
```

```python
data3 = data2[data2['ages'] == 'total']
data3.head()
```

```python
data3['Density'] = data3['population']/data3['area (sq. mi)']
data3.head()
```

```python
data3= data3[data3['year'] == 2010]
data3['Rank'] = data3['Density'].rank(method = 'dense',ascending = False)
data3.head()
```

```python
data4 = data3.sort_values(by = 'Rank', ascending = True)[['state','Density','Rank','year']]
```
